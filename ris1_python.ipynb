```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import colors

M = 20
N = 20
f = 2400000000
c = 30000000000
lam = c/f
d = lam/2
t_t = int(input("Enter the value of incident elevation angle: "))
p_t = int(input("Enter the value of incident azimuth angle: "))
theta_t = t_t*np.pi/180
phi_t = p_t*np.pi/180

t_r = int(input("Enter the value of desired reflected elevation angle: "))
p_r = int(input("Enter the value of desired reflected azimuth angle: "))
theta_r = t_r*np.pi/180
phi_r = p_r*np.pi/180

phase = np.zeros((M,N))
for m in range(M):
  for n in range(N):
    a = -2*(np.pi)*d*((((M+1)/2)-m)*np.cos(phi_r)*np.sin(theta_r) + (n-((N+1)/2))*np.sin(theta_r)*np.sin(phi_r))/lam
    b = 2*(np.pi)*d*((((M+1)/2)-m)*np.cos(phi_t)*np.sin(theta_t) + (n-((N+1)/2))*np.sin(theta_t)*np.sin(phi_t))/lam
    k = (a-b)%(2*np.pi)
    if((k >= 0) and (k < np.pi)):
      phase[m][n] = 0
    elif((k >= np.pi) and (k < 2*np.pi)):
      phase[m][n] = 1

print(phase)

cmap = colors.ListedColormap(['pink', 'green'])
fig, ax = plt.subplots()
ax.matshow(phase, cmap=cmap)

# uncomment the below code to view gridlines on colormap

ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=1)
ax.set_xticks(np.arange(-0.5, M, 1));
ax.set_yticks(np.arange(-0.5, N, 1));

plt.show()
```

### **RIS encoding matrix using *Computer Vision-Aided Reconfigurable Intelligent Surface-Based Beam Tracking: Prototyping and Experimental Results* paper**

**For near field (i.e distance between base station and RIS is less than 800d^2/lambda.(dfeed < 17.14 m ))**

---



```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import colors

M = 20
N = 20

f = 3500000000
c = 300000000

lam = c/f
d = lam/2

t_r = int(input("Enter the value of desired reflected elevation angle: "))
p_r = int(input("Enter the value of desired reflected azimuth angle: "))
theta_r = t_r*np.pi/180
phi_r = p_r*np.pi/180

dfeed = 3.2                       #distance between Base station ans RIS

phase = np.zeros((M,N))
for m in range(M):
  for n in range(N):
    a = -2*(np.pi)*d*((((M+1)/2)-m)*np.cos(theta_r) + (n-(N+1)/2)*np.sin(theta_r)*np.sin(phi_r))/lam
    b = 2*(np.pi)*dfeed/lam - 2*(np.pi)*math.sqrt((d**2)*((((M+1)/2)-m)**2 + (n-(N+1)/2)**2) + (dfeed**2))/lam
    k = (a-b)%(2*np.pi)
    if((k >= 0) and (k < np.pi)):
      phase[m][n] = 0
    elif((k >= np.pi) and (k < 2*np.pi)):
      phase[m][n] = 1

print(phase)

cmap = colors.ListedColormap(['pink', 'green'])
fig, ax = plt.subplots()
ax.matshow(phase, cmap=cmap)

# uncomment the below code to view gridlines on colormap

ax.grid(which='major', axis='both', linestyle='-', color='k', linewidth=1)
ax.set_xticks(np.arange(-0.5, M, 1));
ax.set_yticks(np.arange(-0.5, N, 1));

plt.show()
```

### **Mitigating Quantization Lobes by introducing random error in phase**

**Here we are using 3 different phase randomization matrices to introduce error of 0-60 degrees, 0-120 degrees and 0-180 degrees using uniform random function**

**phase0** is **zero delay** phase array, **phase1** is **0 to 60 degrees** delay phase array, **phase2** is **0 to 120 degrees**, and **phase3** is **0 to 180** degrees.


*For further info, please read the comments in the code(the green colored lines starting with hashtags).*

```python
# importing necessary lobraries
import numpy as np
import random
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import math

# declaring constants i.e. frequency(f), wavelength(lam) and distance between two consecutive RIS elements(d)
# c is the speed of light and k is wavenumber
f = 3500000000
c = 300000000
lam = c/f
k = 2*np.pi/lam
d = lam/2

# declaring the size of array
N = 20
M = 20

# taking the values incident and reflected angles input in degrees
# incident angles
t_t = int(input("Enter the value of incident elevation angle: "))
p_t = int(input("Enter the value of incident azimuth angle: "))
theta_i = t_t*np.pi/180                                               # converting degrees to radians
phi_i = p_t*np.pi/180
# reflected angles
t_r = int(input("Enter the value of desired reflected elevation angle: "))
p_r = int(input("Enter the value of desired reflected azimuth angle: "))
theta_r = t_r*np.pi/180
phi_r = p_r*np.pi/180


# Next, I'm declaring four 2D matrices(initializing with zeros) phase0, phase1, phase2 and phase3 for storing the final phase
# after no randomization, 0-60 degrees randomization, 0-120 degrees randomization and 0-180 degrees randominzation respectively.
phase0 = np.zeros((M,N))
phase_q0 = np.zeros((M,N))                                          # phase_q0, phase_q1, phase_q2 and phase_q3 are the quantized phase delays for the above discusses four cases.
phase_total0 = np.zeros((M,N))                                     # phase_total0, 1, 2 and 3 are the total phase profile of each radiating element for above four cases.

phase1 = np.zeros((M,N))
phase_q1 = np.zeros((M,N))
phase_total1 = np.zeros((M,N))

phase2 = np.zeros((M,N))
phase_q2 = np.zeros((M,N))
phase_total2 = np.zeros((M,N))

phase3 = np.zeros((M,N))
phase_q3 = np.zeros((M,N))
phase_total3 = np.zeros((M,N))

# AF_x, AF_y are real and imaginary components of array factor respectively for all four cases as 0,1,2 and 3
AF_x0 = 0
AF_y0 = 0
AF_x1 = 0
AF_y1 = 0
AF_x2 = 0
AF_y2 = 0
AF_x3 = 0
AF_y3 = 0

# phase_rand0 is the array of zeros only which doesn't introduce any phase delay in any element
phase_rand0 = [[0 for i in range(M)] for j in range(N)]
# phase_rand1 introduces phase delay of 0 to 60 degrees following uniform distribution of random phases
phase_rand1 = [[random.uniform(0, 1)*np.pi/3 for i in range(M)] for j in range(N)]
# phase_rand2 introduces 0 to 120 degrees phase delay
phase_rand2 = [[random.uniform(0, 1)*2*np.pi/3 for i in range(M)] for j in range(N)]                       # values are written in radian here
# phase_rand3 introduces 0 to 180 degrees phase delay
phase_rand3 = [[random.uniform(0, 1)*np.pi for i in range(M)] for j in range(N)]


# phi ranges from -90 to +90 degrees(now it plots 3D graph)
p = np.linspace(-np.pi/2,np.pi/2,100)
# theta is having a range of -90 to +90 degrees(written as pi/2)
t = np.linspace(-np.pi/2,np.pi/2,100)

theta,phi = np.meshgrid(t, p)
# iterating over all the elements in RIS array to assign values in above declared matrices and also the real and imaginary values of array factor
for m in range(M):
  for n in range(N):
    # phase_not is reflected phase
    phase_not = -2*(np.pi)*d*((((M+1)/2)-m)*np.cos(phi_r)*np.sin(theta_r) + (n-((N+1)/2))*np.sin(theta_r)*np.sin(phi_r))/lam
    # phase_illum is incident phase
    phase_illum = 2*(np.pi)*d*((((M+1)/2)-m)*np.cos(phi_i)*np.sin(theta_i) + (n-((N+1)/2))*np.sin(theta_i)*np.sin(phi_i))/lam


    # calculating values of all the 12 previously declared matrices for (m,n) element of array
    # for zero random phase delay
    phase0[m][n] = phase_not - phase_illum - phase_rand0[m][n]
    phase_q0[m][n] = abs(np.pi*(math.floor((phase0[m][n]/np.pi) + 0.5)))
    phase_total0[m][n] = phase_q0[m][n] + phase_illum + phase_rand0[m][n]

    # for 0 to 60 degrees random phase delay
    phase1[m][n] = phase_not - phase_illum - phase_rand1[m][n]
    phase_q1[m][n] = abs(np.pi*(math.floor((phase1[m][n]/np.pi) + 0.5)))
    phase_total1[m][n] = phase_q1[m][n] + phase_illum + phase_rand1[m][n]

    # for 0 to 120 degrees random phase delay
    phase2[m][n] = phase_not - phase_illum - phase_rand2[m][n]
    phase_q2[m][n] = abs(np.pi*(math.floor((phase2[m][n]/np.pi) + 0.5)))
    phase_total2[m][n] = phase_q2[m][n] + phase_illum + phase_rand2[m][n]

    # for 0 to 180 degrees random phase delay
    phase3[m][n] = phase_not - phase_illum - phase_rand3[m][n]
    phase_q3[m][n] = abs(np.pi*(math.floor((phase3[m][n]/np.pi) + 0.5)))
    phase_total3[m][n] = phase_q3[m][n] + phase_illum + phase_rand3[m][n]

    # calculating values of real and imaginary part of array factor for all four cases(AF_x are real parts and AF_y are imaginary parts)
    AF_x0 += np.cos(phase_total0[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))
    AF_y0 += np.sin(phase_total0[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))

    AF_x1 += np.cos(phase_total1[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))
    AF_y1 += np.sin(phase_total1[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))

    AF_x2 += np.cos(phase_total2[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))
    AF_y2 += np.sin(phase_total2[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))

    AF_x3 += np.cos(phase_total3[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))
    AF_y3 += np.sin(phase_total3[m][n] + k*np.sin(theta)*d*(np.cos(phi)*(((M+1)/2)-m) + np.sin(phi)*(n-((N+1)/2))))

# Calculating magnitude of array factor and normalizing it
# for zero random phase delay

AF0 = np.sqrt(AF_x0**2 + AF_y0**2)
max0 = np.amax(AF0)
fig0 = go.Figure(data=[go.Surface(x=theta, y=phi, z=AF0/max0)])
fig0.update_layout(title='For zero randomization',scene = dict(
                    xaxis_title='theta(radians)',
                    yaxis_title='phi(radians)',
                    zaxis_title='Normalized pattern'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig0.show()

# for 0 to 60 degrees random phase delay
AF1 = np.sqrt(AF_x1**2 + AF_y1**2)
max1 = np.amax(AF1)
fig1 = go.Figure(data=[go.Surface(x=theta, y=phi, z=AF1/max1)])
fig1.update_layout(title='For 0 to 60 degrees randomization',scene = dict(
                    xaxis_title='theta(radians)',
                    yaxis_title='phi(radians)',
                    zaxis_title='Normalized pattern'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig1.show()
# for 0 to 120 degrees random phase delay
AF2 = np.sqrt(AF_x2**2 + AF_y2**2)
max2 = np.amax(AF2)
fig2 = go.Figure(data=[go.Surface(x=theta, y=phi, z=AF2/max2)])
fig2.update_layout(title='For 0 to 120 degrees randomization',scene = dict(
                    xaxis_title='theta(radians)',
                    yaxis_title='phi(radians)',
                    zaxis_title='Normalized pattern'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig2.show()
# for 0 to 180 degrees random phase delay
AF3 = np.sqrt(AF_x3**2 + AF_y3**2)
max3 = np.amax(AF3)
fig3 = go.Figure(data=[go.Surface(x=theta, y=phi, z=AF3/max3)])
fig3.update_layout(title='For 0 to 60 degrees randomization',scene = dict(
                    xaxis_title='theta(radians)',
                    yaxis_title='phi(radians)',
                    zaxis_title='Normalized pattern'),
                    width=700,
                    margin=dict(r=20, b=10, l=10, t=10))
fig3.show()

print(max0, max1, max2, max3)
```

As it is clear from the above graphs that for 0 to 180 degrees random delay, the other side lobe or the quantization lobe reaches minimum value, therefore this phase delay is the best for removing it.

Next, I'm showing the resultant binary matrices for zero and 0 to 180 degrees phase delay.

```python
# I'm calculating final values of 0s and 1s that are to be filled in the matrix by assigning 0 to phases 0 to 180 degrees and 1 to 180 to 360 degrees

phase_final0 = np.zeros((M,N))
for m in range(M):
  for n in range(N):
    k = phase0[m][n]%(2*np.pi)# Here, you can change the phase0 matric by any other of the rest two, i.e.phhase1 and phase2 to view them.
    if(( k>= 0) and (k < np.pi)):
      phase_final0[m][n] = 0
    elif((k >= np.pi) and (k < 2*np.pi)):
      phase_final0[m][n] = 1

print("Zero delay phase profile: \n",phase_final0)
print("\n")

phase_final3 = np.zeros((M,N))
for m in range(M):
  for n in range(N):
    k = phase3[m][n]%(2*np.pi)
    if(( k>= 0) and (k < np.pi)):
      phase_final3[m][n] = 0
    elif((k >= np.pi) and (k < 2*np.pi)):
      phase_final3[m][n] = 1

print("0 to 180 degrees delay phase profile: \n",phase_final3)
print("\n")
from matplotlib import colors

cmap = colors.ListedColormap(['pink', 'green'])
figure, axis = plt.subplots(1, 2, figsize = (12,24))
axis[0].matshow(phase_final0, cmap=cmap)
axis[0].set_title("Zero random delay") # change the name of the plot here in case you view any different matrix to avoid confusion
axis[1].matshow(phase_final3, cmap=cmap)
axis[1].set_title("0 to 180 degrees random delay")
# uncomment the below code to view gridlines on colormap

axis[0].grid(which='major', axis='both', linestyle='-', color='k', linewidth=1)
axis[0].set_xticks(np.arange(-0.5, M, 1));
axis[0].set_yticks(np.arange(-0.5, N, 1));
axis[1].grid(which='major', axis='both', linestyle='-', color='k', linewidth=1)
axis[1].set_xticks(np.arange(-0.5, M, 1));
axis[1].set_yticks(np.arange(-0.5, N, 1));

plt.show()
```

**Hence, by using the phase_final3 matrix or the right figure instead of original phase_final0 matrix or the left figure, we can remove the side lobe or the quantization lobe.**
